```python
#creating a development traingle
# A general function that can be used for developing any triangle in actuarial science but impt to remember that the data is arrranged in the ascending order in a sorted manner


def d_triangle(df,accident_yr,reported_yr,parameter):
  columns=[]
  #Setting the names of Columns
  for i in range(1,df[accident_yr].nunique()+1):
    columns.append(12*i)
  k=0                                                                           #This is counter that will ensure the shape of the triangle
  final_list=[]                                                                 #This is the final list of rows in the data frame
  unique=df[accident_yr].nunique()
  for i in range(0,df[accident_yr].nunique()):
    lis=[]
    for j in range(k,k+unique):
      lis.append(df.loc[j].at[parameter])
      k=k+1
    unique=unique-1
    if(len(lis))!=df[accident_yr].nunique():
      g=df[accident_yr].nunique()-len(lis)
      for i in range(0,g):
        lis.append(np.NaN)
    final_list.append(lis)
  bf=pd.DataFrame(final_list,columns=columns)
  bf["Accident Year"]=df[accident_yr].unique()
  first_column = bf.pop('Accident Year')
  bf.insert(0,"Accident Year",first_column)
  return bf





```

```python
# applying the triangle to see the development of reported claims across timeline
#repo_triangle is a dataframe containing reported claims alone

accident_yr=Accident_yr
reported_yr=Reported_yr
parameter=Reported_claims

repo_triangle=pd.DataFrame()
repo_triangle=d_triangle(initial,accident_yr,reported_yr,parameter)
repo_triangle
```

```python
# creating paid_triangle that will contain the paid claims development triangle


accident_yr=Accident_yr
reported_yr=Reported_yr
parameter=Paid_Claims
paid_triangle=pd.DataFrame()
paid_triangle=d_triangle(initial,accident_yr,reported_yr,parameter)
paid_triangle
```

```python
# developing incremental triangle:
def incremental_triangles(cf):

  final=[]
  for i in range(0,len(cf)):
    val=[]
    for j in range(0,len(cf.columns)):
      if j==0 or j==1:
        g=cf.loc[i,cf.columns[j]]
      else:
        if cf.loc[i,cf.columns[j]]!=np.NaN:
          g=cf.loc[i,cf.columns[j]]-cf.loc[i,cf.columns[j-1]]
        else:
          g=np.NaN
      val.append(g)
    final.append(val)
  inc=pd.DataFrame(final,columns=cf.columns)
  return inc

```

```python
paid_incremental_triangle=incremental_triangles(paid_triangle)
paid_incremental_triangle
```

```python
# creating cf that will contain the Case Outstanding development triangle

accident_yr=Accident_yr
reported_yr=Reported_yr
parameter=Case_outstanding
case_out_triangle=pd.DataFrame()
case_out_triangle=d_triangle(initial,accident_yr,reported_yr,parameter)
case_out_triangle
```

```python
#function to calculte paid_tr-on-case ratios
def paid_tr_on_case_ratio(paid_tr,case_out_tr):
  final=[]
  li=list(paid_tr.columns[1:])


  bi=list(paid_tr.columns[2:])
  for i in range(0,len(paid_tr)-1):
    val=[]
    for j in range(0,len(li)-1):
      if paid_tr.loc[i].at[li[j+1]]!=np.NaN and paid_tr.loc[i].at[li[j]]!=np.NaN:
         g=paid_tr.loc[i].at[li[j+1]]-paid_tr.loc[i].at[li[j]]
      else:
        g=np.NaN
      if case_out_tr.loc[i].at[li[j]]!=np.NaN:
        f=case_out_tr.loc[i].at[li[j]]
      else:
        f=np.NaN
      if f!=np.NaN and g!=np.NaN:
        val.append(g/f)
      else:
        val.append(np.NaN)
    final.append(val)
  nf=pd.DataFrame(final,columns=bi)

  return nf





```

```python
paid_ratio_triangle=pd.DataFrame()
paid_ratio_triangle=paid_tr_on_case_ratio(paid_triangle,case_out_triangle)
paid_ratio_triangle
```

```python
#function to calculate Remaining-in-case ratios
def rem_in_case_ratio(case_out_tr):
  li=list(case_out_tr.columns[1:])
  bi=li[1:]
  final=[]
  for i in range(0,len(case_out_tr)-1):
    val=[]
    for j in range(0,len(li)-1):
      if case_out_tr.loc[i].at[li[j]]!=np.NaN and case_out_tr.loc[i].at[li[j+1]]!=np.NaN:
        c=case_out_tr.loc[i].at[li[j+1]]/case_out_tr.loc[i].at[li[j]]
        val.append(c)
      else:
        val.append(np.NaN)
    final.append(val)
  nf=pd.DataFrame(final,columns=bi)
  return nf
```

```python
case_ratio_triangle=pd.DataFrame()
case_ratio_triangle=rem_in_case_ratio(case_out_triangle)
case_ratio_triangle
```

```python
def calc_measures(pf):
  columns=list(pf.columns)
  final=[]

  for i in columns:
    val=[]
    mean=pf[i].mean()
    median=pf[i].median()
    val.append(mean)
    val.append(median)
    final.append(val)
  bf=pd.DataFrame(final)
  bf.columns=["Mean","Median"]
  bf["Months"]=(bf.index+2)*12
  bf.transpose()
  return bf
```

```python
measures_case_out=calc_measures(case_ratio_triangle)
measures_case_out
```

```python
measures_paid_tr=calc_measures(paid_ratio_triangle)
measures_paid_tr
```

```python
#Fill_the_triangles
def fill_triangles_case_outstanding(df,measure,method):
  dup=df
  columns=list(dup.columns)
  for i in range(1,len(dup)):
    for j in range(1,len(columns)):
      if dup.isnull().iloc[i,j]:
        dup.loc[i,columns[j]]=round(measure[method][j-2]*dup.loc[i].at[columns[j-1]],3)
  return dup
nf=case_out=fill_triangles_case_outstanding(case_out_triangle,measures_case_out,"Mean")



```

```python
#paid_tringle_development:
def fill_triangles_paid_triangle(case_out,paid_tr,measure,method):
  columns=list(case_out.columns)
  for i in range(1,len(case_out)):
    for j in range(1,len(columns)):
      if paid_tr.isnull().iloc[i,j]:
        paid_tr.loc[i,columns[j]]=round(measure[method][j-2]*case_out.loc[i].at[columns[j-1]])
  return paid_tr
```
